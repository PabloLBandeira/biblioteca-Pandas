```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
dados = pd.read_excel('/content/golf_df.xlsx')
dados.head()
```

```python
x = pd.get_dummies(dados.iloc[:,:-1]).astype(int)
x.head()
```

```python
y = dados.iloc[:,-1]
y.head()
```

```python
arvore = tree.DecisionTreeClassifier()
arvore.fit(x,y)
```

```python
fig = plt.figure(figsize=(15,10))
_ = tree.plot_tree(arvore, feature_names=x.columns,
                   class_names=['0','1'],
                   filled=True)
```
